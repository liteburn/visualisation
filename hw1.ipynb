```python
import pandas as pd
import altair as alt

df = pd.read_csv("data/temperature.csv", skiprows=1)
df["avg"] = (df["Jan"] + df["Feb"] + df["Mar"] + df["Apr"] + df["May"] + df["Jun"] + df["Jul"] + df["Aug"] + df["Sep"] +
             df["Oct"] + df["Nov"] + df["Dec"]) / 12
df["avg"].round(2)
lst = []
for i in range(1880, 2018):
    if i in [1944, 2014, 2015, 2016, 2017]:
        lst.append(str(i))
    else:
        lst.append("")

lst1 = []
for i in range(1880, 2018):
    if i in [1998, 1904]:
        lst1.append(str(i))
    else:
        lst1.append("")

df["label"] = lst
df["label2"] = lst1
degree_sign = u'\N{DEGREE SIGN}'
chart = alt.Chart(df, title=['Annual Global Surface Temperature,', 'Relative to Late 19th Century Average']).mark_point(
    stroke='black', strokeWidth=0.5, filled=True, fill='#ff3300', size=40).encode(
    alt.X('Year',
          title=None,
          axis=alt.Axis(title='', tickSize=0, labelColor='grey'),
          scale=alt.Scale(domain=[1880, 2017], nice=False)
          ),
    alt.Y('avg',
          title=None,
          axis=alt.Axis(title='',
                        labelExpr=("datum.value+'" + degree_sign + "C'")
                        )
          )
).properties(
    width=800,
    height=500
)

text = chart.mark_text(
    align='left',
    baseline='middle',
    fontSize=10,
    dx=7
).encode(
    text='label'
)
text1 = chart.mark_text(
    align='left',
    baseline='middle',
    fontSize=10,
    dx=-30
).encode(
    text='label2'
)
line = alt.Chart(pd.DataFrame({'y': [0.0]})).mark_rule().encode(y='y')

axis_text = alt.Chart(pd.DataFrame([{'x': 2010, 'y': 0}])).mark_text(
    text=['HOTTER THAN THE', '1880-1899 AVERAGE', '', 'COLDER'],
    dy=-25
).encode(
    x='x:Q',
    y='y:Q'
)

alt.layer(chart, text, text1, line, axis_text).configure_point(
    color='gray',
    fill='#E55B13',
    strokeWidth=0.8
).configure_title(
    anchor='start',
    dx=50,
    dy=50,
    fontWeight=800
).configure_axis(
    labelColor='gray',
    ticks=False,
    domain=False,
).configure_view(
    strokeOpacity=0
)





```
